 #  Detection of toxic comments using Support Vector Machine (SVM) 
 

Sentiment Analysis is a process of extracting contextual meaning of words by applying natural language processing, text analysis, computational linguistics, and biometrics.
A basic task in sentiment analysis is classifying the polarity of a given text whether the expressed opinion is posititve, negative or neutral.

For example:
*     "You look beautiful" = Positive
*     "The food is terrible and really expensive" = Negative
*     "I am hungry" = Neutral

SVM is a type of supervised machine learning algorithm. The dataset used here is "Jigsaw Multilingual Toxic Comment Classification" provided by Google and a manually created "Toxic Messages" dataset.

```python
stopword = stopwords.words('english')
remove = ["no", "nor", "not", "doesn't", "hasn't", "haven't", "isn't", "mustn't","needn't", "shouldn't", "shan't","wasn't", "won't", "weren't", "wouldn't"]
for x in remove:
    stopword.remove(x)
    
print(stopword)
```

```python
train1['comment_text_vector'] = train1['comment_text']
train1['comment_text_vector'] = train1['comment_text_vector'].apply(lambda x: re.sub('[^a-zA-Z]',' ',str(x)))
train1['comment_text_vector'] = train1['comment_text_vector'].apply(lambda x: x.lower())
train1['comment_text_vector'] = train1['comment_text_vector'].apply(lambda x: x.split())
train1['comment_text_vector'] = train1['comment_text_vector'].apply(lambda x: [item for item in x if item not in stopword])
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pickle

filename = 'TextModel_SVM.sav'
clf = pickle.load(open(filename, 'rb'))
vec = CountVectorizer(decode_error="replace",vocabulary=pickle.load(open(r'Vocabulary.sav', "rb")))
hindi_toxic = pd.read_csv('Hindi.csv')
```

```python
@anvil.server.callable

def line_processing(input_data):

    text = []
    text.append(input_data)
    score = 0
    keyword = re.sub('[^a-zA-Z]',' ',text[0])
    for i in keyword.split(' '):
        for j in hindi_toxic['Words']:
            if j == i:
                score = 1
                break
    if score == 1:
        print('Score: ', score)
        return([score], 'Not Okay')
    else:
        X = vec.transform(text)
        score = clf.predict(X)
        print('Score', score)
        if score == [1]:
            return(score, 'Not Okay')
        else:
            return(score, 'Okay')
    
```
