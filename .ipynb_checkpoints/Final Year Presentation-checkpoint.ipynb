# Text Processing

## 1. Importing Libraries

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
nltk.download('stopwords')
```

## 2.Importing Datasets

```python
df1 = pd.read_csv(r'jigsaw-final.csv')
df1.head()
```

```python
df2 = pd.read_excel(r'toxicmessage.xlsx', encoding = 'latin-1')
df2.head()
```

## 3. Data Cleansing

```python
train1 = pd.concat([df1,df2] , axis = 0)
train1.reset_index(inplace=True)
train1['comment_text']
```

```python
train1['TOXIC'].value_counts()
```

```python
train1.shape
```

```python
sns.countplot(train1['TOXIC'])
```

```python
text = train1['comment_text'].loc[train1['TOXIC'] == 0]
from wordcloud import WordCloud, STOPWORDS
stopwords_data = set(STOPWORDS)

plt.figure(figsize=(10,10))
plt.title('Common words in non toxic sentences')
plt.imshow(WordCloud(stopwords=stopwords_data).generate(str(text)))
```

```python
text = train1['comment_text'].loc[train1['TOXIC'] == 1]
from wordcloud import WordCloud, STOPWORDS
stopwords_data = set(STOPWORDS)

plt.figure(figsize=(10,10))
plt.title('Common words in toxic sentences')
plt.imshow(WordCloud(stopwords=stopwords_data).generate(str(text)))
```

```python
stopword = stopwords.words('english')
remove = ["no", "nor", "not", "doesn't", "hasn't", "haven't", "isn't", "mustn't","needn't", "shouldn't", "shan't","wasn't", "won't", "weren't", "wouldn't"]
for x in remove:
    stopword.remove(x)
    
print(stopword)
```

```python
train1['comment_text_vector'] = train1['comment_text']
train1['comment_text_vector'] = train1['comment_text_vector'].apply(lambda x: re.sub('[^a-zA-Z]',' ',str(x)))
train1['comment_text_vector'] = train1['comment_text_vector'].apply(lambda x: x.lower())
train1['comment_text_vector'] = train1['comment_text_vector'].apply(lambda x: x.split())
train1['comment_text_vector'] = train1['comment_text_vector'].apply(lambda x: [item for item in x if item not in stopword])
```

```python
wordnet_lemmatizer = WordNetLemmatizer()
train1['comment_text_vector'] = train1['comment_text_vector'].apply(lambda x: [wordnet_lemmatizer.lemmatize(i) for i in x ])
```

```python
train1['comment_text'][0]
```

```python
train1['comment_text_vector'][0]
```

```python
count_vect = TfidfVectorizer()

X = train1['comment_text_vector']
X_new = [[' '.join(i)] for i in X]
X_new = [i for x in X_new for i in x]
X_train_tf = count_vect.fit_transform(X_new)

dense = X_train_tf.todense()
dense = dense.tolist()
dense = pd.DataFrame(dense)
```
